```python
# Q3: How do the distributions of bill_length_mm and bill_depth_mm differ between the three penguin species? 
# Analyze the skewness and kurtosis of each feature for different species.
grouped_data = df.groupby('species')

#skewness for bill_length_mm and bill_depth_mm
skewness_bill_length = grouped_data['bill_length_mm'].skew()
skewness_bill_depth = grouped_data['bill_depth_mm'].skew()

# kurtosis for bill_length_mm and bill_depth_mm 
kurtosis_bill_length = grouped_data['bill_length_mm'].apply(pd.Series.kurtosis)
kurtosis_bill_depth = grouped_data['bill_depth_mm'].apply(pd.Series.kurtosis)


(skewness_bill_length, kurtosis_bill_length, skewness_bill_depth, kurtosis_bill_depth)
```

```python
#Q4: Identify which features in the dataset have outliers. Provide the method used to detect them and visualize the outliers.
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Function to identify outliers using the IQR method
def find_outliers(feature):
    Q1 = df[feature].quantile(0.25)
    Q3 = df[feature].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df[(df[feature] < lower_bound) | (df[feature] > upper_bound)]
    return outliers

# Identifying outliers for numerical features
numerical_features = ['bill_length_mm', 'bill_depth_mm', 'flipper_length_mm', 'body_mass_g']

outliers = {}
for feature in numerical_features:
    outliers[feature] = find_outliers(feature)
    print(f"Outliers for {feature}:")
    print(outliers[feature])

# Visualize outliers with boxplots
plt.figure(figsize=(12, 8))
for i, feature in enumerate(numerical_features, 1):
    plt.subplot(2, 2, i)
    
    # Check if the feature has non-empty data to plot
    if not df[feature].dropna().empty:
        sns.boxplot(x=df[feature])
        plt.title(f'Outliers in {feature}')
    else:
        plt.text(0.5, 0.5, f"No data to display for {feature}", 
                 horizontalalignment='center', verticalalignment='center',
                 fontsize=12, color='red')
    
plt.tight_layout()
plt.show()


```

```python
#Q6: Use bill_length_mm vs bill_depth_mm and plot 7 different graphs to visualize them. 
plt.figure(figsize=(20, 15))

# Scatter plot
plt.subplot(3, 3, 1)
sns.scatterplot(x='bill_length_mm', y='bill_depth_mm', hue='species', data=df)
plt.title('Scatter Plot')

# Hexbin plot
plt.subplot(3, 3, 2)
plt.hexbin(df['bill_length_mm'], df['bill_depth_mm'], gridsize=30, cmap='Blues')
plt.colorbar()
plt.title('Hexbin Plot')

# KDE plot
plt.subplot(3, 3, 3)
sns.kdeplot(x='bill_length_mm', y='bill_depth_mm', data=df, fill=True)
plt.title('KDE Plot')

# Violin plot
plt.subplot(3, 3, 4)
sns.violinplot(x='species', y='bill_length_mm', data=df)
plt.title('Violin Plot - Bill Length')

# Box plot
plt.subplot(3, 3, 5)
sns.boxplot(x='species', y='bill_depth_mm', data=df)
plt.title('Box Plot - Bill Depth')

# Pair plot
plt.subplot(3, 3, 6)
sns.pairplot(df[['bill_length_mm', 'bill_depth_mm', 'species']], hue='species')
plt.title('Pair Plot')

# Joint plot
plt.subplot(3, 3, 7)
sns.jointplot(x='bill_length_mm', y='bill_depth_mm', data=df, kind='reg')
plt.title('Joint Plot')

plt.tight_layout()
plt.show()
```
